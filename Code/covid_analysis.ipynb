```python
location_variant_summary = df.groupby(['location', 'variant']).agg({
     'num_sequences' : 'sum',
     'perc_sequences' : 'mean',
     'num_sequences_total' : 'sum'
})
location_variant_summary
```

# Vizualisation

## Time-Series Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data = location_data, x='date', y='num_sequences', hue='variant')
plt.title(f'Trends of Varients in {location} Over time')
plt.xlabel('Date')
plt.ylabel('Num of Sequences')
plt.legend(title='Varient')
plt.show()
```

## Geographical Insights

```python
plt.figure(figsize=(29,6))
sns.barplot(data=location_summary, x='location', y='num_sequences', palette='viridis')
plt.xticks(rotation= 45, ha='right')
plt.title('Total no of Sequences by Location', fontsize=14)
plt.xlabel('location', fontsize=12)
plt.ylabel('No of Sequences', fontsize=12)
plt.show()
```

## Proportional Analysis

```python
plt.figure(figsize=(22, 6))
sns.barplot(data=varient_summary, x='date', y='perc_sequences', hue='variant', palette='viridis')
plt.title('Percentage Contribution of Variants Over Time')
plt.xticks(rotation= 45, ha='right')
plt.xlabel('Date')
plt.ylabel('Percentage of Sequences')
plt.legend(title='Variant')
plt.show()
```

## Interactive Visualization

# Advance Analysis (Forecast)

```python
df['growth_rate'] = df.groupby(['location', 'variant'])['num_sequences'].pct_change()
df['growth_rate']
```

```python
# %pip install statsmodels
from statsmodels.tsa.arima.model import ARIMA
```

```python
variant = 'Omnicorn'
variant_data = df[(df['location'] == location) & (df['variant'] == 'Omicron')].set_index('date')
variant_data.head()
```

```python
model = ARIMA(variant_data['num_sequences'], order=(1,1,1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=606)
print(forecast)
```
